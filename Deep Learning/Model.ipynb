```python
#Membangun network dengan keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(8,input_dim=4,activation='relu')) #8 = jumlah neuron , imput deminsion 4
model.add(Dense(8,input_dim=4,activation='relu')) #8 = jumlah neuron , imput deminsion 4
model.add(Dense(3,activation='softmax')) #output ada 3 neuron, karena ada 3 target
model.compile(loss = 'categorical_crossentropy', optimizer='adam', metrics=['accuray'])
```
